# Carregando os Dados

Informações do Dataset

This is a dataset of 60,000 28x28 grayscale images of the 10 digits, along with a test set of 10,000 images


- Returns

Tuple of NumPy arrays: (x_train, y_train), (x_test, y_test).


```python
from tensorflow import keras
import cv2
import numpy as np
import matplotlib.pyplot as plt
from random import randint
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
(X_train_raw, y_train_raw), (X_test_raw, y_test_raw) = keras.datasets.fashion_mnist.load_data()
```

```python
print("X_train: ", X_train_raw.shape)
print("y_train: ", y_train_raw.shape)
print("X_test:  ", X_test_raw.shape)
print("y_test:  ", y_test_raw.shape)
```

```python
X = []
for image in range(X_train_raw.shape[0]):
  img_res = cv2.resize(X_train_raw[image], (14,14))
  X.append(img_res)
```

```python
X = np.array(X)
print(X.shape)
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
f, ax = plt.subplots(5,5)
f.subplots_adjust(0,0,1.5,1.5)
for i in range(0,5,1):
  for j in range(0,5,1):
    rnd_number = randint(0,len(X))
    ax[i,j].imshow(X[rnd_number], cmap=plt.cm.binary)
    ax[i,j].set_title(class_names[y_train_raw[rnd_number]])
    ax[i,j].axis('off')
```

# Pré-processamento dos Dados

```python
X_train, X_val, y_train, y_val = train_test_split(
     X, y_train_raw, stratify=y_train_raw, test_size=0.33, random_state=42)
```

```python
hot = OneHotEncoder()
y_train = hot.fit_transform(y_train.reshape(-1,1)).toarray()
y_val = hot.transform(y_val.reshape(-1,1)).toarray()
```

```python
X_train = X_train / 255
X_val = X_val / 255
```

```python
print(X_train.shape)
print(X_val.shape)
print(y_train.shape)
print(y_val.shape)
```

Distribuição das Classes

```python
import seaborn as sns
sns.countplot(y_train_raw);
```

#Modelo Deep Learning

## Carregando as Bibliotecas

```python
from tensorflow.python.keras.layers import Input, Dense, Activation, Dropout, Flatten
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.callbacks import EarlyStopping

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

## Modelo 5

Redução das dimensões das imagens (14x14)

### Definição da Arquitetura

```python
rede = Sequential()

rede.add(Flatten())
rede.add(Dense(25,  input_dim = X_train.shape[1]*X_train.shape[2]))
rede.add(Activation('relu'))
rede.add(Dense(64))
rede.add(Activation('relu'))
rede.add(Dropout(0.5))
rede.add(Dense(64))
rede.add(Activation('relu'))
rede.add(Dense(128))
rede.add(Activation('relu'))
rede.add(Dropout(0.5))
rede.add(Dense(128))
rede.add(Activation('relu'))
rede.add(Dense(10))
rede.add(Activation('softmax'))
```

```python
es = EarlyStopping(monitor = 'val_loss', 
                   min_delta = 0.001, 
                   patience = 10, 
                   verbose = 1, 
                   mode = 'auto')
```

```python
rede.compile(loss = 'categorical_crossentropy', 
              optimizer = 'Adam', 
              metrics=['accuracy'])
```

### Treinamento do Modelo

```python
hist = rede.fit(X_train, 
                y_train, 
                epochs = 50,
                batch_size=50,
                verbose = 1,
                validation_data = (X_val, y_val),
                callbacks = [es])
```

### Avaliação do Modelo

```python
pred = rede.predict(X_val)

accuracy_score(y_val.argmax(1), pred.argmax(1))
```

```python
loss = hist.history['loss']
val_loss = hist.history['val_loss']
epochs = range(len(loss))
plt.figure()
plt.plot(epochs, loss, 'bo', label = 'Training loss')
plt.plot(epochs, val_loss, 'b', label = 'Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```
